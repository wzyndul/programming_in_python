# First point of assignment 1

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
headlines = ("sex","length [mm]","diameter [mm]","height [mm]","whole weight [g]","shucked weight [g]", "viscera weight [g]", "shell weight [g]", "rings")
```

```python
snails_df = pd.read_csv("data\data.csv", names = headlines)
```

```python
male_counter = snails_df['sex'].value_counts()['M']
female_counter = snails_df['sex'].value_counts()['F']
infant_counter = snails_df['sex'].value_counts()['I']
all_snails_counter = len(snails_df['sex'])

qualitative_data = {'category name': ['Male', 'Infant', 'Female'],
                    'count': [male_counter, infant_counter,female_counter],
                    '%':[round(male_counter/all_snails_counter*100,2), round(infant_counter/all_snails_counter*100,2),round(female_counter/all_snails_counter*100,2)]
}

qualitative_df = pd.DataFrame(qualitative_data)
qualitative_df = qualitative_df.set_index('category name', drop='False')
```

```python
display(qualitative_df)
```

```python
display(quantitative_df)
```

```python
#Bar plot 
fig = plt.figure(figsize = (10,5))
x_sex = list(qualitative_df.index.values)
y_sex = list(qualitative_df['count'])
plt.bar(x_sex,y_sex, color='maroon', width = 0.4)
```

```python
#Histograms

rows_names = ["length [mm]", "diameter [mm]", "height [mm]", "whole weight [g]", "shucked weight [g]", "viscera weight [g]", "shell weight [g]", "rings"]

plt.figure(figsize=(7,20))

counter1 = 1
for i in rows_names:
    plt.subplot(4,2,counter1)
    plt.hist(snails_df[i],color='maroon', bins=20, alpha=0.8)
    plt.title(i)
    counter1 += 1

```

```python
#Scatters
rows_names = ["length [mm]", "diameter [mm]", "height [mm]", "whole weight [g]", "shucked weight [g]", "viscera weight [g]", "shell weight [g]", "rings"]

plt.figure(figsize=(15,70))

counter2 = 1
counter3 = 1

for i in rows_names:
    rows_names2 = rows_names[counter3:len(rows_names)]
    for j in rows_names2:
        plt.subplot(14,2,counter2)
        plt.scatter(snails_df[i],snails_df[j])
        plt.title(i + "-" + j)
        counter2 += 1
    counter3 += 1
```

```python
#Linear correlation matrix
cor_matrix = snails_df.iloc[:,1:].corr()
display(cor_matrix)
```

```python
#Heatmap representing a linear correlation matrix
heatmap = sns.heatmap(cor_matrix, annot=True)
display(heatmap)
```

```python
#linear regression plot with the two quantitative variables that are most strongly linearly correlated
reg_plot = sns.regplot(snails_df.iloc[:,1:], x="length [mm]", y="diameter [mm]", line_kws=dict(color="r"))
display(reg_plot)
```
